## Limpieza adicional

```python
# añadir uid las filas con un modulo que genera id 

import hashlib #para operaciones criptograficas

uids = (spotify
        .apply(lambda row: hashlib.md5(bytes(row['track_id'].encode())), axis=1) #axis=1 filas / axis=0 column
        .apply(lambda hash_object: hash_object.hexdigest())
       )
spotify['uid'] = uids
spotify.set_index('uid', inplace=True) #Inplace modificar directamente la tabla, no retorna otra df

spotify
```

```python
stripped_body = (spotify
                .apply(lambda row:row['album_name'],axis=1)
                .apply(lambda album_name:list(album_name))
               )
stripped_body
```
